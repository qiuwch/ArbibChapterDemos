# Generate sinusoid stimuli

Sinusoid $ I(\mathbf{x}) = A \cos(\mathbf{\omega} \mathbf{x} + \rho) $

$ A $ is the amplitude, $ \rho $ is the phase, and $ \mathbf{\omega} $ is the frequency.

```python
def genSinusoid(sz, A, omega, rho): # omega < 0.2
    # Generate Sinusoid grating
    # sz: size of generated image (width, height)
    [x, y] = meshgrid(arange(0, sz[0]), arange(0, sz[1]))
    stimuli = A * cos(omega[0] * x  + omega[1] * y + rho)
    return stimuli

theta = pi/4
omega = [cos(theta), sin(theta)]
sinusoidParam = {'A':1, 'omega':omega, 'rho':0, 'sz':(32,32)}
myimshow(genSinusoid(**sinusoidParam))
```

# Generate gabor filter
A general type of Gabor filter[1] can be defined:  
$$ g(x,y;\lambda,\theta,\psi,\sigma,\gamma) = \exp\left(-\frac{x'^2+\gamma^2y'^2}{2\sigma^2}\right)\exp\left(i\left(2\pi\frac{x'}{\lambda}+\psi\right)\right) $$

Here we implement the type of Gabor filter which satisfies the neurophysiological constraints for simple cells[2]: 

$$ \psi(x, y, \omega_0, \theta) = \left[ \frac{\omega_0}{\sqrt{2\pi} K} \exp \{ - \frac{\omega_0^2}{8 K^2}(4(x \cos\theta + y  \sin\theta))^2 + ( -x \sin \theta + y \cos \theta)^2) \} \right]\left[ \exp \{ i\omega_0(x \cos\theta + y \sin \theta ) \} - \exp(-\frac{K^2}{2})  \right] $$

[1] https://en.wikipedia.org/wiki/Gabor_filter  
[2] Lee, Tai Sing. "Image representation using 2D Gabor wavelets." Pattern Analysis and Machine Intelligence, IEEE Transactions on 18.10 (1996): 959-971.

```python
def genGabor(sz, omega, theta, func=cos):
    radius = (int(sz[0]/2.0), int(sz[1]/2.0))
    [x, y] = meshgrid(range(-radius[0], radius[0]+1), range(-radius[1], radius[1]+1))

    x1 = x * cos(theta) + y * sin(theta)
    y1 = -x * sin(theta) + y * cos(theta)
    K = pi
    
    gauss = omega / (sqrt(2*pi) * K) * exp(- omega**2 / (8*K**2) * ( 4 * x1**2 + y1**2))
    # figure(); imshow(gauss)
    sinusoid = func(omega * x1) - exp(-K**2 / 2)
    # figure(); imshow(sinusoid)
    gabor = gauss * sinusoid
    return gabor
        
myimshow(genGabor((256,256), 0.1, pi/4))
```

# Generate gabor filter bank

```python
theta = arange(0, pi, pi/4)
omega = arange(0.2, 0.6, 0.1)
params = [(t,o) for o in omega for t in theta]
sinFilterBank = []
cosFilterBank = []
gaborParams = []
for (theta, omega) in params:
    gaborParam = {'omega':omega, 'theta':theta, 'sz':(128, 128)}
    sinGabor = genGabor(func=sin, **gaborParam)
    cosGabor = genGabor(func=cos, **gaborParam)
    sinFilterBank.append(sinGabor)
    cosFilterBank.append(cosGabor)
    gaborParams.append(gaborParam)

figure()
n = len(sinFilterBank)
for i in range(n):
    subplot(4,4,i+1)
    title(r'$\theta$={theta:.2f}$\omega$={omega}'.format(**gaborParams[i]))
    myimshow(sinFilterBank[i])

figure()
for i in range(n):
    subplot(4,4,i+1)
    title(r'$\theta$={theta:.2f}$\omega$={omega}'.format(**gaborParams[i]))
    myimshow(cosFilterBank[i])
```

# Apply filter bank to zebra image

```python
from skimage.color import rgb2gray
from scipy.signal import convolve2d
zebra = rgb2gray(imread('data/Zebra_running_Ngorongoro.jpg')) # range(0->1)
myimshow(zebra)
sinGabor = sinFilterBank[1]
myimshow(sinGabor)
%time res = convolve2d(zebra, sinGabor) # Will take about one minute
figure(); myimshow(res); title('response')
```

# Quadrature pair, simple/complex cell

```python
theta = pi/4
sinGabor = genGabor((129,129), 0.4, theta, sin)
cosGabor = genGabor((129,129), 0.4, theta, cos)
figure(); 
subplot(121); myimshow(sinGabor); subplot(122); myimshow(cosGabor)

theta = pi/4 + pi
sinusoid = genSinusoid((256,256), 1, (omega*sin(theta), omega*cos(theta)), 0)
figure(); myimshow(sinusoid); title('Stimuli')

response = convolve2d(sinusoid, sinGabor, mode='valid')
response2 = convolve2d(sinusoid, cosGabor, mode='valid')

figure(); 
subplot(121); imshow(response, vmin=0); title('Response of sin gabor(simple cell)')
subplot(122); imshow(response**2 + response2**2, vmin=0); title('Resp. of complex cell')
```

**Complex cell is tuned to frequency and less sensitive to phase change**

# Demo: Gaussian, Laplacian of Gaussian

```python
# Utility function
def surf(X, Y, Z, **kargs): 
    # Plot 3D data as surface, similar to surf(X,Y,Z) of http://www.mathworks.com/help/matlab/ref/surf.html
    from mpl_toolkits.mplot3d import Axes3D
    fig = figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, **kargs)
    
sigma = 20
from mpl_toolkits.mplot3d import Axes3D

[X, Y] = meshgrid(arange(-100, 101), arange(-100, 101))
Z = 1/(sqrt(2.0 * pi) * sigma) * exp(-(X**2+Y**2)/(2.0*sigma**2))

dx = np.roll(Z, 1, axis=1) - Z
dx2 = np.roll(dx, 1, axis=1) - dx

dy = np.roll(Z, 1, axis=0) - Z
dy2 = np.roll(dy, 1, axis=0) - dy

LoG = -(dx2+dy2)

surf(X, Y, Z, alpha=0.3)
title('Gaussian')

surf(X, Y, dx + dy, alpha=0.3)
title('First order derivative')

surf(X, Y, LoG, alpha=0.3)
title('Second order derivative (Laplacian of Gaussian)')
```
