# Generate sinusoid stimuli

Sinusoid $ I(\mathbf{x}) = A \cos(\mathbf{\omega} \mathbf{x} + \rho) $

$ A $ is the amplitude, $ \rho $ is the phase, and $ \mathbf{\omega} $ is the frequency.

```python
def genSinusoid(sz, A, omega, rho): # omega < 0.2
    # Generate Sinusoid grating
    # sz: size of generated image (width, height)
    [x, y] = meshgrid(arange(0, sz[0]), arange(0, sz[1]))
    stimuli = A * cos(omega[0] * x  + omega[1] * y + rho)
    return stimuli

theta = pi/4
omega = [cos(theta), sin(theta)]
sinusoidParam = {'A':1, 'omega':omega, 'rho':0, 'sz':(32,32)}
myimshow(genSinusoid(**sinusoidParam))
```

# Generate gabor filter
A general type of Gabor filter[1] can be defined:  
$$ g(x,y;\lambda,\theta,\psi,\sigma,\gamma) = \exp\left(-\frac{x'^2+\gamma^2y'^2}{2\sigma^2}\right)\exp\left(i\left(2\pi\frac{x'}{\lambda}+\psi\right)\right) $$
[1] https://en.wikipedia.org/wiki/Gabor_filter  

Here we implement the type of Gabor filter which satisfies the neurophysiological constraints for simple cells:   
$$ \psi (x; \omega, \theta, K) = \frac{\omega^2}{ 4 \pi K^2} \exp  \{-(\omega^2/8K^2)[4(x\cdot(cos\theta, sin\theta))^2 + (x \cdot ( -sin \theta, cos \theta))^2]\} \times \exp \{ iwx \cdot (cos\theta, sin\theta) \} exp(K^2/2) $$

```python
def genGabor(sz, omega, theta, func=cos, K=pi):
    radius = (int(sz[0]/2.0), int(sz[1]/2.0))
    [x, y] = meshgrid(range(-radius[0], radius[0]+1), range(-radius[1], radius[1]+1))

    x1 = x * cos(theta) + y * sin(theta)
    y1 = -x * sin(theta) + y * cos(theta)
    
    gauss = omega**2 / (4*pi * K**2) * exp(- omega**2 / (8*K**2) * ( 4 * x1**2 + y1**2))
#     myimshow(gauss)
    sinusoid = func(omega * x1) * exp(K**2 / 2)
#     myimshow(sinusoid)
    gabor = gauss * sinusoid
    return gabor
        
g = genGabor2((256,256), 0.3, pi/4, func=cos)
myimshow(g)
mean(g)
```

# Generate gabor filter bank

```python
theta = arange(0, pi, pi/4)
omega = arange(0.2, 0.6, 0.1)
params = [(t,o) for o in omega for t in theta]
sinFilterBank = []
cosFilterBank = []
gaborParams = []
for (theta, omega) in params:
    gaborParam = {'omega':omega, 'theta':theta, 'sz':(128, 128)}
    sinGabor = genGabor(func=sin, **gaborParam)
    cosGabor = genGabor(func=cos, **gaborParam)
    sinFilterBank.append(sinGabor)
    cosFilterBank.append(cosGabor)
    gaborParams.append(gaborParam)

figure()
n = len(sinFilterBank)
for i in range(n):
    subplot(4,4,i+1)
    title(r'$\theta$={theta:.2f}$\omega$={omega}'.format(**gaborParams[i]))
    axis('off'); imshow(sinFilterBank[i])

figure()
for i in range(n):
    subplot(4,4,i+1)
    title(r'$\theta$={theta:.2f}$\omega$={omega}'.format(**gaborParams[i]))
    axis('off'); imshow(cosFilterBank[i])
```

# Demo: Gaussian, Laplacian of Gaussian

```python
# Utility function
def surf(X, Y, Z, **kargs): 
    # Plot 3D data as surface, similar to surf(X,Y,Z) of http://www.mathworks.com/help/matlab/ref/surf.html
    from mpl_toolkits.mplot3d import Axes3D
    fig = figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, **kargs)
    
sigma = 20
from mpl_toolkits.mplot3d import Axes3D

[X, Y] = meshgrid(arange(-100, 101), arange(-100, 101))
Z = 1/(sqrt(2.0 * pi) * sigma) * exp(-(X**2+Y**2)/(2.0*sigma**2))

dx = np.roll(Z, 1, axis=1) - Z
dx2 = np.roll(dx, 1, axis=1) - dx

dy = np.roll(Z, 1, axis=0) - Z
dy2 = np.roll(dy, 1, axis=0) - dy

LoG = -(dx2+dy2)

surf(X, Y, Z, alpha=0.3)
title('Gaussian')

surf(X, Y, dx + dy, alpha=0.3)
title('First order derivative')

surf(X, Y, LoG, alpha=0.3)
title('Second order derivative (Laplacian of Gaussian)')
```
