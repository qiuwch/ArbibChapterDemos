```python
# Show image and edge labeling
# id = '100075'
# id = '35010' # butterfly
# id = '97017' # building
id = '41004' # deer

def loadData(id):
    from skimage.color import rgb2gray
    edgeMap = imread('boundaryMap/' + id + '.bmp'); edgeMap = edgeMap[:,:,0]
    im = imread('BSR/BSDS500/data/images/train/' + id + '.jpg')
    grayIm = rgb2gray(im)
    return [grayIm, edgeMap]

[im, edgeMap] = loadData(id)
figure(); myimshow(edgeMap); title('Edge')
figure(); myimshow(im); title('Image')
figure(); myimshow(im * (edgeMap==0))
```

```python
def dIdx(im):
    # First order derivative
    dx1 = abs(np.roll(im, 1, axis=1) - im)
    dx2 = abs(np.roll(im, -1, axis=1) - im)
    dy1 = abs(np.roll(im, 1, axis=0) - im)
    dy2 = abs(np.roll(im, -1, axis=0) - im)
    dx = np.maximum(dx1, dx2)
    dy = np.maximum(dy1, dy2)
    dIm = np.maximum(dx, dy)
    return dIm

def dIdx2(im):
    dx = np.roll(im, 1, axis=1) - im
    dx2 = np.roll(dx, 1, axis=1) - dx
    return dx2

def dgIdx(im):
    from scipy import ndimage
    pixelRadius = 3; nderivative = 1;
    dgauss = ndimage.gaussian_filter(im, sigma = pixelRadius/2, order=(nderivative, nderivative))
    return dgauss

dx = dIdx(im)
dx2 = dIdx2(im)
dgI = dgIdx(im)
    
figure(); myimshow(abs(dx)); title(r'$ \frac{dI}{dx} $')
figure(); myimshow(abs(dx2)); title(r'$ \frac{d^2 I}{dx^2} $')
figure(); myimshow(abs(dgI)); title(r'$ \frac{d G*I}{dx} $')
figure(); imshow(abs(dx)>0.2)
# showEdge(abs(dgauss), edgeMap)

# def showEdge(im, edgeMap):
#     # draw edge pixel
#     im = im * (edgeMap != 0)
#     figure(); myimshow(im); title('Highlight edge')
```

```python
def kde(x):
    from scipy.stats import gaussian_kde
    f = gaussian_kde(x, bw_method=0.01 / x.std(ddof=1))
    return f.evaluate
    
def ponEdge(im, edgeMap):
    # Compute on edge histogram
    # im is filtered image
    
    # Convert edge map to pixel index
    flattenEdgeMap = edgeMap.flatten()
    edgeIdx = [i for i in range(len(flattenEdgeMap)) if flattenEdgeMap[i]]
    
    # find edge pixel in 3x3 region, Shift the edge map a bit
    [offx, offy] = np.meshgrid(np.arange(-1,2), np.arange(-1,2)); offx = offx.flatten(); offy = offy.flatten()
    maxVal = np.copy(im)
    for i in range(9):
        im1 = np.roll(im, offx[i], axis=1) # x axis
        im1 = np.roll(im1, offy[i], axis=0) # y axis    
        maxVal = np.maximum(maxVal, im1)

    # figure(); imshow(maxVal)
    vals = maxVal.flatten()
    onEdgeVals = vals[edgeIdx]
    
    bins = np.linspace(0,0.5, 100)
    [n, bins] = np.histogram(onEdgeVals, bins=bins)
    n = n+1 # Avoid divide by zero
    # p = n / sum(n)

    pon = kde(onEdgeVals)

    return [n, bins, pon]


def poffEdge(im, edgeMap):
    flattenEdgeMap = edgeMap.flatten()
    noneEdgeIdx = [i for i in range(len(flattenEdgeMap)) if not flattenEdgeMap[i]]
    
    vals = im.flatten()
    offEdgeVals = vals[noneEdgeIdx] 

    bins = np.linspace(0,0.5, 100)
    n, bins = np.histogram(offEdgeVals, bins=bins)

    n = n+1 # prevent divide by zero error
    p = n / sum(n)
    
    poff = kde(offEdgeVals)
    
    return [n, bins, poff]

dx = abs(dIdx(im))
[n1, bins, pon] = ponEdge(dx, edgeMap)
figure(); # Plot on edge



[n2, bins, poff] = poffEdge(dx, edgeMap)
figure(); # Plot on edge
title('On/Off edge')
plot((bins[:-1] + bins[1:])/2, n1.astype(float)/sum(n1), label="pon")
plot((bins[:-1] + bins[1:])/2, n2.astype(float)/sum(n2), label="poff")
legend()

figure()
plot(bins, pon(bins), color='blue', alpha=0.5, lw=3, label="pon")
plot(bins, poff(bins), color='blue', alpha=0.5, lw=3, label="poff")
legend()
```
